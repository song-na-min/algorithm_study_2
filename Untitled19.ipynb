```python
#중복순열 사용
# 중복순열과 eval 함수, 복사를 사용

from itertools import permutations

def calcul_with_permutations(n,num_arr,cal_arr):
  #3,4,5=> ((3+4)+5)
  calcul_arr=[]
  for _ in range(n-1):
    calcul_arr.append("(")
  calcul_arr.append(num_arr[0])
  print(calcul_arr)
  cal=[]
  calculator=["+","-","*","//"]
  for c in range(4):
    if cal_arr[c]!=0:
      for _ in range(cal_arr[c]):
        cal.append(calculator[c])
  print(cal)
  min_value=100000
  max_value=-100000
  for i in permutations(cal,n-1):
    #print(i)
    calcul_test_arr=[]
    result=0
    calcul_test_arr=calcul_arr[:]
    #for _ in range(n-1):
    #  calcul_test_arr.append("(")
    #calcul_test_arr.append(num_arr[0])
    #print(calcul_test_arr)
    #print(calcul_arr)
    for k in range(n-1):
      calcul_test_arr.append(i[k])
      calcul_test_arr.append(num_arr[k+1])
      calcul_test_arr.append(")")
    #print(calcul_test_arr)
    result=eval("".join(calcul_test_arr))
    #print(result)
    min_value=min(min_value,result)
    max_value=max(max_value,result)
  print(min_value,max_value)
```

#### 중복순열에서 문제점

음수 나누기에서 예외처리가 불가능하다 실제 예제 3번의 경우 최댓값이 54가 아닌 48이 나오는데 이는 파이썬과 c++14에서의 나눗셈 결과가 다르기 때문이다. 따라서 eval 함수를 사용하지 않고 앞에서부터 계산 결과를 확인하는 것이 좋을 것 같다.
https://soyoja.com/88

dfs의 답지를 보고 수정 
연산을 할 때 int(-48/5)를 하면 원하는 결과를 얻을 수 있다 그래서 이전과 달리 숫자와 연산 하나씩 확인하면서 앞뒤로 ()를 붙이도록 수정하였다.

```python
#중복순열 사용
# 중복순열과 eval 함수, 복사를 사용

from itertools import permutations

def calcul_with_permutations_int(n,num_arr,cal_arr):
  #3,4,5=> ((3+4)+5)
  cal=[]
  calculator=["+","-","*","/"]
  for c in range(4):
    if cal_arr[c]!=0:
      for _ in range(cal_arr[c]):
        cal.append(calculator[c])
  print(cal)
  min_value=100000
  max_value=-100000
  for i in permutations(cal,n-1):
    calcul_test_arr=[num_arr[0]]
    for k in range(n-1):
      if i[k]=="/":
        calcul_test_arr.insert(0,"int(")
        #print(calcul_test_arr)
      else:
        calcul_test_arr.insert(0,"(")
      calcul_test_arr.append(i[k])
      calcul_test_arr.append(num_arr[k+1])
      calcul_test_arr.append(")")
    #print(calcul_test_arr)
    result=eval("".join(calcul_test_arr))
    #print(result)
    min_value=min(min_value,result)
    max_value=max(max_value,result)
  print(min_value,max_value)
```

```python
n=6
num_arr=["1","2","3","4","5","6"]#(((((1+2)+3)+4)+5)+6)
cal_arr=[2,1,1,1]
calcul_with_permutations_int(n,num_arr,cal_arr)
print(-48/5)
print(-48//5)
print(-(48//5))
print(eval("int(-48/5)"))
```

### dfs 사용

사실 dfs

```python
n=6
num_arr=[1,2,3,4,5,6]#(((((1+2)+3)+4)+5)+6)
cal_arr=[2,1,1,1]
min_val=1e9
max_val=-1e9

def dfs(i,now):
  global min_val,max_val
  #모든 연산자를 다 확인하면 계산해서 update
  if i==n:
    min_val=min(min_val,now)
    max_val=max(max_val,now)      
  else:
    if cal_arr[0]>0:
      cal_arr[0]-=1
      #연산자를 넣고 dfs
      dfs(i+1,now+num_arr[i])
      #백트래킹
      cal_arr[0]+=1
    if cal_arr[1]>0:
      cal_arr[1]-=1
      dfs(i+1,now-num_arr[i])
      cal_arr[1]+=1
    if cal_arr[2]>0:
      cal_arr[2]-=1
      dfs(i+1,now*num_arr[i])
      cal_arr[2]+=1
    if cal_arr[3]>0:
      cal_arr[3]-=1
      dfs(i+1,int(now/num_arr[i]))
      cal_arr[3]+=1
dfs(1,num_arr[0])
print(max_val)
print(min_val)
```
