#### 중복순열에서 문제점

음수 나누기에서 예외처리가 불가능하다 실제 예제 3번의 경우 최댓값이 54가 아닌 48이 나오는데 이는 파이썬과 c++14에서의 나눗셈 결과가 다르기 때문이다. 따라서 eval 함수를 사용하지 않고 앞에서부터 계산 결과를 확인하는 것이 좋을 것 같다.

dfs의 답지를 보고 수정 
연산을 할 때 int(-48/5)를 하면 원하는 결과를 얻을 수 있다 그래서 이전과 달리 숫자와 연산 하나씩 확인하면서 앞뒤로 ()를 붙이도록 수정하였다.

```python
#중복순열 사용
# 중복순열과 eval 함수, 복사를 사용

from itertools import permutations

def calcul_with_permutations_int(n,num_arr,cal_arr):
  #3,4,5=> ((3+4)+5)
  cal=[]
  calculator=["+","-","*","/"]
  for c in range(4):
    if cal_arr[c]!=0:
      for _ in range(cal_arr[c]):
        cal.append(calculator[c])
  print(cal)
  min_value=100000
  max_value=-100000
  for i in permutations(cal,n-1):
    calcul_test_arr=[num_arr[0]]
    for k in range(n-1):
      if i[k]=="/":
        calcul_test_arr.insert(0,"int(")
        #print(calcul_test_arr)
      else:
        calcul_test_arr.insert(0,"(")
      calcul_test_arr.append(i[k])
      calcul_test_arr.append(num_arr[k+1])
      calcul_test_arr.append(")")
    #print(calcul_test_arr)
    result=eval("".join(calcul_test_arr))
    #print(result)
    min_value=min(min_value,result)
    max_value=max(max_value,result)
  print(min_value,max_value)
```

```python
n=6
num_arr=["1","2","3","4","5","6"]#(((((1+2)+3)+4)+5)+6)
cal_arr=[2,1,1,1]
calcul_with_permutations_int(n,num_arr,cal_arr)
print(-48/5)
print(-48//5)
print(-(48//5))
print(eval("int(-48/5)"))
```

### dfs 사용

사실 dfs

```python
n=6
num_arr=[1,2,3,4,5,6]#(((((1+2)+3)+4)+5)+6)
cal_arr=[2,1,1,1]
min_val=1e9
max_val=-1e9

def dfs(i,now):
  global min_val,max_val
  #모든 연산자를 다 확인하면 계산해서 update
  if i==n:
    min_val=min(min_val,now)
    max_val=max(max_val,now)      
  else:
    if cal_arr[0]>0:
      cal_arr[0]-=1
      #연산자를 넣고 dfs
      dfs(i+1,now+num_arr[i])
      #백트래킹
      cal_arr[0]+=1
    if cal_arr[1]>0:
      cal_arr[1]-=1
      dfs(i+1,now-num_arr[i])
      cal_arr[1]+=1
    if cal_arr[2]>0:
      cal_arr[2]-=1
      dfs(i+1,now*num_arr[i])
      cal_arr[2]+=1
    if cal_arr[3]>0:
      cal_arr[3]-=1
      dfs(i+1,int(now/num_arr[i]))
      cal_arr[3]+=1
dfs(1,num_arr[0])
print(max_val)
print(min_val)
```

```python

def is_ok(n,arr):
  for i in range(n):
    for j in range(n):
      if arr[i][j]=="S" and search(n,arr,i,j)==False:
        return False
  return True
def search(n,arr,i,j):
  for h in range(i,0,-1):
    if arr[h][j]=="T":
      return False
    if arr[h][j]=="O":
      break
  for h in range(i,n):
    if arr[h][j]=="T":
      return False
    if arr[h][j]=="O":
      break
  for h in range(j,0,-1):
    if arr[i][h]=="T":
      return False
    if arr[i][h]=="O":
      break
  for h in range(j,n):
    if arr[i][h]=="T":
      return False
    if arr[i][h]=="O":
      break
  return True
result="No"
def avoid_surveillance(n,arr,count):
  global result
  if count==3:
    temp=[[0]*n for _ in range(n)]
    for i in range(n):
      for j in range(n):
        temp[i][j]=arr[i][j]
    if is_ok(n,temp)==True:
      result="YES"
      #print(temp)
    return
  for i in range(n):
    for j in range(n):
      if arr[i][j]=="X":
        arr[i][j]="O"
        count+=1
        avoid_surveillance(n,arr,count)
        arr[i][j]="X"
        count-=1
  return result
```

## 21 인구이동 ...

```python
from collections import deque
n,l,r=3,5,10

arr=[[10,15,20],
          [20,30,25],
          [40,22,10]]

dx=[-1,0,1,0]
dy=[0,-1,0,1]

def bfs(x,y,index):
        #united : 연합에 있는 배열의 x,y좌표 저장
        #=> 각 칸의 인구수를 바꿀 수 있도록
        united=[]
        united.append((x,y))
        q=deque()
        q.append((x,y))
        union[x][y]=index
        summary=arr[x][y]
        count=1

        while q:
            x,y=q.popleft()
            #상하좌우 검사
            for i in range(4):
                nx=x+dx[i]
                ny=y+dy[i]
                #만약 범위 내에있고, 연합이 정해지지 않았으면
                if 0<=nx<n and 0<=ny<n and union[nx][ny]==-1:
                    # 인구 차이가 l이상 r 이하이면 
                    if l<=abs(arr[nx][ny]-arr[x][y])<=r:
                        #연합에 추가
                        q.append((nx,ny))
                        #연합에 추가한 것을 배열에 반영(다음에 검사 안함)
                        union[nx][ny]=index
                        #합 추가
                        summary+=arr[nx][ny]
                        count+=1
                        united.append((nx,ny))
        #연합에 추가한 것들의 x,y좌표를 가져와서
        for i,j in united:
            #해당 칸의 인구수를 변경
            arr[i][j]=summary//count
        return count
total_count=0
while True:
        #초기화
        union=[[-1]*n for _ in range(n)]
        index=0
        for i in range(n):
            for j in range(n):
                # 아직 정해지 연합이 없는 경우 주변 확인
                if union[i][j]==-1:
                    bfs(i,j,index)
                    index+=1

        if index==n*n:
            break
        total_count+=1
print(total_count)

```

```python

from collections import deque

def get_next_pos(pos,board):
    next_pos=[]
    pos=list(pos)
    pos1_x,pos1_y,pos2_x,pos2_y=pos[0][0],pos[0][1],pos[1][0],pos[1][1]
    dx=[-1,1,0,0]
    dy=[0,0,-1,1]
    for i in range(4):
        pos1_next_x,pos1_next_y,pos2_next_x,pos2_next_y=pos1_x+dx[i],pos1_y+dy[i],pos2_x+dx[i],pos2_y+dy[i]
        if board[pos1_next_x][pos1_next_y]==0 and board[pos2_next_x][pos2_next_y]==0:
            next_pos.append({(pos1_next_x,pos1_next_y),(pos2_next_x,pos2_next_y)})
    if pos1_x==pos2_x:
        for i in [-1,1]:
            if board[pos1_x+i][pos1_y]==0 and board[pos2_x+i][pos2_y]==0:
                next_pos.append({(pos1_x,pos1_y),(pos1_x+i,pos1_y)})
                next_pos.append({(pos2_x,pos2_y),(pos2_x+i,pos2_y)})
    elif pos1_y==pos2_y:
        for i in [-1,1]:
            if board[pos1_x][pos1_y+i]==0 and board[pos2_x][pos2_y+i]==0:
                next_pos.append({(pos1_x,pos1_y),(pos1_x,pos1_y+i)})
                next_pos.append({(pos2_x,pos2_y),(pos2_x,pos2_y+i)})
    return next_pos

def solution(board):
    answer = 0
    n=len(board)
    new_board=[[1]*(n+2) for _ in range(n+2)]
    for i in range(n):
        for j in range(n):
            new_board[i+1][j+1]=board[i][j]
    q=deque()
    visited=[]
    pos={(1,1),(1,2)}
    q.append((pos,0))
    visited.append(pos)
    while q:
        pos,cost=q.popleft()
        if (n,n) in pos:
            return cost
        for next_pos in get_next_pos(pos,new_board):
            if next_pos not in visited:
                q.append((next_pos,cost+1))
                visited.append(next_pos)
    return 0
```
